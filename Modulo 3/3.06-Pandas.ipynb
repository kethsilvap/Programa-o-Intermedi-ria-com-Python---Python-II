````python
import pandas as pd

# Carregar o dataset correto
wc = pd.read_csv("matches_1930_2022.csv")

# 🔧 Padronizar nomes das colunas (caso venham em inglês)
wc = wc.rename(columns={
    'HomeTeam': 'time_1',
    'AwayTeam': 'time_2',
    'HomeGoals': 'gols_1',
    'AwayGoals': 'gols_2',
    'Date': 'data',
    'Year': 'ano',
    'Attendance': 'comparecimento',
    'HostCountry': 'país_sede',
    'Stage': 'rodada'
})

# 🔤 Converter para string (somente se existir)
cols_string = [
    'time_1', 'time_2', 'país_sede', 'resultado',
    'gols_1_detalhes', 'gols_2_detalhes',
    'gols_1_contra', 'gols_2_contra',
    'gols_1_penalti', 'gols_2_penalti',
    'cartao_vermelho_1', 'cartao_vermelho_2',
    'cartao_amarelo_1', 'cartao_amarelo_2',
    'copa'
]

for col in cols_string:
    if col in wc.columns:
        wc[col] = wc[col].astype("string")

# 🔢 Converter para inteiro
cols_int = ['gols_1', 'gols_2', 'ano', 'comparecimento']

for col in cols_int:
    if col in wc.columns:
        wc[col] = wc[col].astype("Int64")  # seguro para valores nulos

# 📅 Converter data
if 'data' in wc.columns:
    wc['data'] = pd.to_datetime(wc['data'], errors='coerce')

# 📊 Converter para categoria
if 'rodada' in wc.columns:
    wc['rodada'] = wc['rodada'].astype("category")

# 💾 Salvar arquivo final
wc.to_csv('wc_formatado.csv', index=False)

# 🔍 Conferir resultado
wc.info()
````

#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7
Colombia        Feminina    7
```

```
país            copa        num_copas
Brazil          Masculina   21
Germany         Masculina   21
Argentina       Masculina   20
England         Masculina   20
Mexico          Masculina   20
```

````python
import pandas as pd

# Carregar dataset
wc = pd.read_csv("matches_1930_2022.csv")

# 🔍 Detectar nomes reais das colunas automaticamente
colunas = wc.columns

time1 = 'HomeTeam' if 'HomeTeam' in colunas else colunas[0]
time2 = 'AwayTeam' if 'AwayTeam' in colunas else colunas[1]
ano = 'Year' if 'Year' in colunas else colunas[2]

# Criar coluna de copa
if 'Tournament' in colunas:
    wc['copa'] = wc['Tournament'].apply(
        lambda x: 'Feminina' if 'Women' in str(x) else 'Masculina'
    )
else:
    wc['copa'] = wc[ano].apply(
        lambda x: 'Feminina' if x >= 1991 else 'Masculina'
    )

# 🔁 Transformar jogos em participações
times_1 = wc[[time1, ano, 'copa']].rename(columns={time1: 'país'})
times_2 = wc[[time2, ano, 'copa']].rename(columns={time2: 'país'})

participacoes = pd.concat([times_1, times_2])

# Remover duplicatas
participacoes = participacoes.drop_duplicates()

# Contar copas por país
participacao = (
    participacoes
    .groupby(['país', 'copa'])[ano]
    .nunique()
    .reset_index(name='num_copas')
)

# Top 5 Feminina
top_fem = (
    participacao[participacao['copa'] == 'Feminina']
    .sort_values(by='num_copas', ascending=False)
    .head(5)
)

# Top 5 Masculina
top_masc = (
    participacao[participacao['copa'] == 'Masculina']
    .sort_values(by='num_copas', ascending=False)
    .head(5)
)

# Resultado
print(top_fem)
print()
print(top_masc)
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
import pandas as pd

# Carregar dataset
wc = pd.read_csv("matches_1930_2022.csv")

# 🔍 Detectar colunas automaticamente
colunas = wc.columns

time1 = 'HomeTeam' if 'HomeTeam' in colunas else colunas[0]
time2 = 'AwayTeam' if 'AwayTeam' in colunas else colunas[1]

gols1 = 'HomeGoals' if 'HomeGoals' in colunas else colunas[2]
gols2 = 'AwayGoals' if 'AwayGoals' in colunas else colunas[3]

# Garantir que gols são numéricos
wc[gols1] = pd.to_numeric(wc[gols1], errors='coerce')
wc[gols2] = pd.to_numeric(wc[gols2], errors='coerce')

# 🔁 Criar estrutura unificada
gols_casa = wc[[time1, gols1]].rename(columns={
    time1: 'país',
    gols1: 'gols'
})

gols_fora = wc[[time2, gols2]].rename(columns={
    time2: 'país',
    gols2: 'gols'
})

# 🔗 Juntar tudo
gols_total = pd.concat([gols_casa, gols_fora])

# 📊 Somar gols por país
gols = (
    gols_total
    .groupby('país')['gols']
    .sum()
    .reset_index(name='total_gols')
)

# 🔝 Ordenar
gols = gols.sort_values(by='total_gols', ascending=False)

# 🖨️ Resultado
print(gols)
````

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import pandas as pd
import ast

# Carregar dataset
wc = pd.read_csv("matches_1930_2022.csv")

colunas = wc.columns

# 🔍 Detectar colunas principais
time1 = 'HomeTeam' if 'HomeTeam' in colunas else colunas[0]
time2 = 'AwayTeam' if 'AwayTeam' in colunas else colunas[1]

# 🔍 Detectar colunas de cartão amarelo automaticamente
cart1 = None
cart2 = None

for col in colunas:
    if 'yellow' in col.lower() and ('home' in col.lower() or '1' in col.lower()):
        cart1 = col
    elif 'yellow' in col.lower() and ('away' in col.lower() or '2' in col.lower()):
        cart2 = col

# 🧠 Função para contar cartões
def contar_cartoes(valor):
    if pd.isna(valor):
        return 0
    try:
        lista = ast.literal_eval(valor)
        return len(lista)
    except:
        return 0

# Aplicar função
wc['num_cartoes_1'] = wc[cart1].apply(contar_cartoes)
wc['num_cartoes_2'] = wc[cart2].apply(contar_cartoes)

# 🔁 Separar casa e fora
casa = wc[[time1, 'num_cartoes_1']].rename(columns={
    time1: 'país',
    'num_cartoes_1': 'cartoes'
})

fora = wc[[time2, 'num_cartoes_2']].rename(columns={
    time2: 'país',
    'num_cartoes_2': 'cartoes'
})

# 🔗 Juntar tudo
cartoes_total = pd.concat([casa, fora])

# 📊 Somar por país
resultado = (
    cartoes_total
    .groupby('país')['cartoes']
    .sum()
    .reset_index(name='cartoes_amarelos')
)

# 🔝 Ordenar
resultado = resultado.sort_values(by='cartoes_amarelos', ascending=False)

# 🖨️ Resultado final
print(resultado)

# 🥇 País com mais cartões
print("\nPaís com mais cartões amarelos:")
print(resultado.iloc[0])
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
import pandas as pd

# Carregar dataset
wc = pd.read_csv("matches_1930_2022.csv")

# 🔍 Garantir que colunas existem (ou detectar automaticamente)
colunas = wc.columns

gols_cols = [
    col for col in colunas
    if 'gols' in col.lower() or 'goal' in col.lower()
]

# Filtrar só as colunas relevantes
gols_cols = [col for col in gols_cols if any(x in col.lower() for x in ['detalhes', 'penalti', 'detail', 'penalty'])]

# 🧠 Dicionário de contagem
contagem = {}

# 🔁 Percorrer colunas e linhas
for col in gols_cols:
    for valor in wc[col].dropna():
        
        # separar gols da partida
        gols = str(valor).split('|')
        
        for g in gols:
            # extrair nome do jogador
            if '·' in g:
                nome = g.split('·')[0].strip()
            else:
                nome = g.strip()
            
            # ignorar vazios
            if nome == '':
                continue
            
            # contar
            contagem[nome] = contagem.get(nome, 0) + 1

# 📊 Converter para DataFrame
gols_df = pd.DataFrame(
    list(contagem.items()),
    columns=['jogador(a)', 'num_gols']
)

# 🔝 Ordenar e pegar top 10
gols_df = gols_df.sort_values(by='num_gols', ascending=False).head(10)

# 🖨️ Resultado
print(gols_df)
````
